```python

import emnist
import numpy as np

train_images, train_labels = emnist.extract_training_samples('letters')
train_images = train_images.astype(np.float32)
train_labels = np.array(train_labels).astype(np.int32)
## h√§r testar man datan 
test_images, test_labels = emnist.extract_test_samples('letters')
test_images = test_images.astype(np.float32)
test_labels = np.array(test_labels).astype(np.int32)
print(" training images:", train_images.shape[0])
print(" test images:", test_images.shape[0])

```

```python
import matplotlib.pyplot as plt

# Plot a few sample images
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(8, 8))
for i, ax in enumerate(axes.flat):
    ax.imshow(train_images[i], cmap='gray')
    ax.set_title(chr(train_labels[i]+96))  # convert label to character
    ax.axis('off')

plt.show()
```
